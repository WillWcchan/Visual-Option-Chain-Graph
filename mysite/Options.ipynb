```python
import requests
import json
from datetime import datetime

response = requests.get('https://sandbox.tradier.com/v1/markets/timesales',
            params={'symbol': 'DIS200807C00145000', 'interval': '5min', 'start': '2020-08-06 09:30', 'end': '2020-08-06 16:00', 'session_filter': 'all'},
            headers={'Authorization': 'Bearer ICTXIFyPoZuZMLnnjxURtZpXTXAR', 'Accept': 'application/json'}
)
json_response = response.json()
print(json.dumps(json_response, indent = 2))
```

```python
from datetime import datetime
# print(json_response["series"]["data"][0])
# for data in list(json_response["series"]["data"]):
#     print(datetime.fromtimestamp(data['timestamp']))
timestamps = [str(datetime.fromtimestamp(data['timestamp']).strftime("%-I:%-M")) for data in json_response["series"]["data"]]
print(timestamps)
price = [data['price'] for data in json_response["series"]["data"]]

plt.plot(timestamps,price)
plt.xlabel("Time")
plt.xticks(rotation=45)
plt.ylabel("Price")
plt.show()
```
